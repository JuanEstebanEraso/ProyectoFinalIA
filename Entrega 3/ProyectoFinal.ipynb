##Funciones para leer metadatos y parsear nombre

````python
# Extrae un entero del nombre "Video 1.mp4" -> video_id = 1
pat_simple = re.compile(r'video\s*(?P<id>\d+)', re.IGNORECASE)

def parse_filename(fname: str):
    name = os.path.splitext(os.path.basename(fname))[0]
    m = pat_simple.search(name)
    video_id = int(m.group("id")) if m else None
    return {
        "video_id": video_id,
        "subject": None,
        "actividad": None,
        "vista": None,
        "repeticion": None
    }

def read_video_meta(path: str):
    cap = cv2.VideoCapture(path)
    if not cap.isOpened():
        return None
    fps   = float(cap.get(cv2.CAP_PROP_FPS) or 0)
    nfr   = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)
    w     = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) or 0)
    h     = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) or 0)
    cap.release()

    dur_s = (nfr / fps) if fps > 0 else None
    size_mb = os.path.getsize(path) / 1024 / 1024
    return {
        "fps": round(fps, 2) if fps else None,
        "frames": nfr if nfr > 0 else None,
        "dur_s": round(dur_s, 2) if dur_s else None,
        "width": w, "height": h,
        "size_mb": round(size_mb, 2)
    }
````

##Construir catálogo

````python
rows = []
for root, _, files in os.walk(BASE_DIR):
    for f in files:
        if f.lower().endswith(".mp4"):
            p = os.path.join(root, f)
            meta = read_video_meta(p)
            if meta is None:
                continue
            info = parse_filename(f)
            rows.append({
                "file": f,
                "path": p,
                **info,
                **meta,
                "rel_dir": os.path.relpath(root, BASE_DIR)
            })

df = pd.DataFrame(rows)

# Orden natural por video_id y luego por nombre
if "video_id" in df.columns:
    df = df.sort_values(by=["video_id","file"], na_position="last").reset_index(drop=True)
else:
    df = df.sort_values(by=["file"]).reset_index(drop=True)

print("Total de videos:", len(df))
df.head(10)
````

##Generar plantillas de etiquetado

````python
import os
import pandas as pd
import numpy as np

# === rutas (deben coincidir con lo que ya usaste) ===
BASE_DIR   = "/content/drive/MyDrive/ProyectoIA/Videos"
CATALOGO   = os.path.join(BASE_DIR, "catalogo_videos.csv")
LABELS_DIR = os.path.join(BASE_DIR, "labels")
os.makedirs(LABELS_DIR, exist_ok=True)

# === etiquetas permitidas para esta entrega ===
ALLOWED = ["sentado", "de_pie", "caminando", "transicion"]

# === secuencia base (proporciones de la duración total; suman 1.0) ===
# Nota: 'transicion' se usa para levantarse, giros y sentarse.
seq = [
    ("sentado",    0.12, "inicio sentado"),
    ("transicion", 0.05, "levantarse"),
    ("de_pie",     0.03, "breve de_pie antes de caminar"),
    ("caminando",  0.30, "caminar hacia la cámara"),
    ("transicion", 0.05, "giro 1"),
    ("caminando",  0.30, "caminar de regreso a la silla"),
    ("transicion", 0.05, "giro 2"),
    ("transicion", 0.05, "sentarse"),
    ("sentado",    0.05, "fin sentado"),
]

# mínimos y redondeo
MIN_SEG = 0.30   # duración mínima por segmento en segundos
ROUND   = 2      # decimales para exportar

FORCE_OVERWRITE = True

df_cat = pd.read_csv(CATALOGO)

def plantilla_path(file_name):
    base = os.path.splitext(file_name)[0]
    return os.path.join(LABELS_DIR, f"{base}_labels.csv")

generadas, omitidas = 0, 0

for _, r in df_cat.iterrows():
    file = r["file"]
    dur  = float(r["dur_s"]) if pd.notna(r["dur_s"]) else None
    if not dur or dur <= 0:
        print(f"[AVISO] Duración inválida en {file}, se omite.")
        continue

    out_csv = plantilla_path(file)
    if (not FORCE_OVERWRITE) and os.path.exists(out_csv):
        omitidas += 1
        continue

    # 1) Proporciones → segundos (asegurando mínimos)
    props = np.array([p for _, p, _ in seq], dtype=float)
    props = props / props.sum()                 # normaliza
    secs  = props * dur

    # aplica mínimos y reescala para que sumen 'dur'
    secs = np.maximum(secs, MIN_SEG)
    secs = secs * (dur / secs.sum())

    # 2) construye intervalos acumulando tiempos
    rows = []
    start = 0.0
    for (label, _, note), seg_dur in zip(seq, secs):
        end = start + float(seg_dur)
        rows.append({
            "start_s": round(start, ROUND),
            "end_s":   round(min(end, dur), ROUND),  # cap por seguridad
            "label":   label,
            "note":    note,  # columna opcional para tu referencia
        })
        start = end
    # fuerza fin EXACTO al total de video
    rows[-1]["end_s"] = round(dur, ROUND)

    tpl = pd.DataFrame(rows, columns=["start_s","end_s","label","note"])
    tpl.to_csv(out_csv, index=False)
    generadas += 1

print(f"Plantillas generadas: {generadas} | omitidas (existentes y FORCE_OVERWRITE=False): {omitidas}")
print("Carpeta de etiquetas:", LABELS_DIR)
!ls -lh "$LABELS_DIR" | head -n 25
````

##Validador de etiquetas

````python
BASE_DIR   = "/content/drive/MyDrive/ProyectoIA/Videos"
CATALOGO   = os.path.join(BASE_DIR, "catalogo_videos.csv")
LABELS_DIR = os.path.join(BASE_DIR, "labels")

ALLOWED = ["sentado", "de_pie", "caminando", "transicion"]
TOL = 0.20  # tolerancia en segundos para suma de intervalos vs duración de video

df_cat = pd.read_csv(CATALOGO)

def plantilla_path(file_name):
    base = os.path.splitext(file_name)[0]
    return os.path.join(LABELS_DIR, f"{base}_labels.csv")

def load_labels_for(file_name):
    p = plantilla_path(file_name)
    if not os.path.exists(p):
        return None
    d = pd.read_csv(p)
    # nos aseguramos de que existan columnas mínimas
    for c in ["start_s","end_s","label"]:
        assert c in d.columns, f"Falta columna {c} en {p}"
    return d

def check_intervals(df_lab):
    ok, msgs = True, []
    d = df_lab.sort_values("start_s").reset_index(drop=True)
    # valida etiquetas
    invalid = sorted(set(d["label"]) - set(ALLOWED))
    if invalid:
        ok = False
        msgs.append(f"Etiquetas inválidas: {invalid}")
    # valida límites y solapes
    for i in range(len(d)):
        if d.loc[i,"end_s"] <= d.loc[i,"start_s"]:
            ok = False
            msgs.append(f"Intervalo con end<=start en fila {i}")
        if i>0 and d.loc[i,"start_s"] < d.loc[i-1,"end_s"]:
            ok = False
            msgs.append(f"Solape entre filas {i-1} y {i}")
    return ok, msgs, d

resumen, errores = [], []
for _, r in df_cat.iterrows():
    fn  = r["file"]
    dur = float(r["dur_s"]) if pd.notna(r["dur_s"]) else None
    labs = load_labels_for(fn)
    if labs is None:
        errores.append((fn, ["No existe archivo de etiquetas"]))
        continue

    ok, msgs, d = check_intervals(labs)
    suma = (d["end_s"] - d["start_s"]).clip(lower=0).sum()
    dur_ok = (dur is not None) and (abs(suma - dur) <= TOL)
    if not dur_ok:
        ok = False
        msgs.append(f"Suma de intervalos={suma:.2f}s != duracion video={dur:.2f}s (±{TOL}s)")

    if not ok:
        errores.append((fn, msgs))
    else:
        resumen.append({
            "file": fn,
            "dur_video_s": round(dur,2),
            "dur_etiquetas_s": round(suma,2),
            "num_intervalos": len(d)
        })

res_df = pd.DataFrame(resumen).sort_values("file")
print("Archivos correctos:", len(res_df), " | Con errores:", len(errores))
display(res_df.head(30))

if errores:
    print("\n--- Detalles de errores ---")
    for fn, msgs in errores:
        print(f"* {fn}")
        for m in msgs:
            print("  -", m)
````

````python
import matplotlib.pyplot as plt

BASE_DIR   = "/content/drive/MyDrive/ProyectoIA/Videos"
CATALOGO   = os.path.join(BASE_DIR, "catalogo_videos.csv")
LABELS_DIR = os.path.join(BASE_DIR, "labels")

df_cat = pd.read_csv(CATALOGO)

def plantilla_path(file_name):
    base = os.path.splitext(file_name)[0]
    return os.path.join(LABELS_DIR, f"{base}_labels.csv")

def load_labels_for(file_name):
    p = plantilla_path(file_name)
    if not os.path.exists(p):
        return None
    return pd.read_csv(p)

def stack_all_labels(df_catalog):
    rows = []
    for _, r in df_catalog.iterrows():
        labs = load_labels_for(r["file"])
        if labs is None:
            continue
        d = labs.copy()
        d["file"]     = r["file"]
        d["video_id"] = r["video_id"]
        d["dur_seg"]  = (d["end_s"] - d["start_s"]).clip(lower=0)
        rows.append(d)
    if not rows:
        return pd.DataFrame(columns=["file","video_id","start_s","end_s","label","dur_seg"])
    return pd.concat(rows, ignore_index=True)

labs_all = stack_all_labels(df_cat)
if len(labs_all)==0:
    print("Aún no hay etiquetas completadas.")
else:
    # Cobertura por etiqueta (segundos totales)
    cobertura = labs_all.groupby("label")["dur_seg"].sum().sort_values(ascending=False)
    print("Cobertura (segundos totales) por etiqueta:")
    print(cobertura)

    # Conteo por etiqueta y por video
    cobertura_por_video = labs_all.groupby(["file","label"])["dur_seg"].sum().reset_index()
    display(cobertura_por_video.head(30))

    # Histogramas de duración de segmentos por etiqueta
    for et in cobertura.index.tolist():
        d = labs_all[labs_all["label"]==et]["dur_seg"]
        if len(d)==0:
            continue
        plt.figure(figsize=(6,4))
        d.plot(kind="hist", bins=15, title=f"Duración de segmentos — {et}")
        plt.xlabel("segundos"); plt.ylabel("conteo")
        plt.show()
````

##Preparar salidas y calcular tablas clave

````python
import os, pandas as pd, numpy as np

BASE_DIR   = "/content/drive/MyDrive/ProyectoIA/Videos"
CATALOGO   = os.path.join(BASE_DIR, "catalogo_videos.csv")
LABELS_DIR = os.path.join(BASE_DIR, "labels")
EDA_DIR    = os.path.join(BASE_DIR, "eda_outputs")
os.makedirs(EDA_DIR, exist_ok=True)

df_cat = pd.read_csv(CATALOGO)

def lab_path(fn):
    base = os.path.splitext(fn)[0]
    return os.path.join(LABELS_DIR, f"{base}_labels.csv")

# ---- 1) Resumen del catálogo ----
n_videos  = len(df_cat)
dur_stats = df_cat["dur_s"].describe().to_frame(name="dur_s")
fps_stats = df_cat["fps"].describe().to_frame(name="fps")

# resoluciones
df_cat["res"] = df_cat["width"].astype(str)+"x"+df_cat["height"].astype(str)
res_counts = df_cat["res"].value_counts().rename_axis("res").reset_index(name="count")

catalog_summary = pd.DataFrame({
    "n_videos":[n_videos],
    "dur_min_s":[df_cat["dur_s"].min()],
    "dur_max_s":[df_cat["dur_s"].max()],
    "dur_mean_s":[df_cat["dur_s"].mean()],
    "fps_min":[df_cat["fps"].min()],
    "fps_max":[df_cat["fps"].max()],
    "fps_mean":[df_cat["fps"].mean()],
})

# ---- 2) Cargar etiquetas y apilar ----
rows = []
for _, r in df_cat.iterrows():
    p = lab_path(r["file"])
    if not os.path.exists(p):
        continue
    d = pd.read_csv(p)
    d["file"] = r["file"]
    d["video_id"] = r["video_id"]
    d["dur_seg"] = (d["end_s"] - d["start_s"]).clip(lower=0)
    rows.append(d)
labs_all = pd.concat(rows, ignore_index=True) if rows else pd.DataFrame()

# totales por etiqueta
label_totals = labs_all.groupby("label")["dur_seg"].sum().sort_values(ascending=False).reset_index()

# cobertura por video y etiqueta (segundos y % del video)
cov_video = labs_all.groupby(["file","label"])["dur_seg"].sum().reset_index()
cov_video = cov_video.merge(df_cat[["file","dur_s"]], on="file", how="left")
cov_video["pct_video"] = 100 * cov_video["dur_seg"] / cov_video["dur_s"]

# pivot “etiquetas x video” en %
cov_pivot_pct = cov_video.pivot_table(index="file", columns="label", values="pct_video", fill_value=0).reset_index()

# ---- 3) Guardar CSVs ----
catalog_summary.to_csv(os.path.join(EDA_DIR,"catalog_summary.csv"), index=False)
res_counts.to_csv(os.path.join(EDA_DIR,"resolutions_counts.csv"), index=False)
dur_stats.to_csv(os.path.join(EDA_DIR,"duration_stats_full.csv"))
fps_stats.to_csv(os.path.join(EDA_DIR,"fps_stats_full.csv"))
label_totals.to_csv(os.path.join(EDA_DIR,"label_totals_seconds.csv"), index=False)
cov_video.to_csv(os.path.join(EDA_DIR,"label_coverage_by_video.csv"), index=False)
cov_pivot_pct.to_csv(os.path.join(EDA_DIR,"label_coverage_pct_pivot.csv"), index=False)

print("Listo. Tablas del EDA guardadas en:", EDA_DIR)
````

### **C5 — Extracción de landmarks por segmento, con caché**

````python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False,
                    model_complexity=1,
                    min_detection_confidence=0.5,
                    min_tracking_confidence=0.5)

def extract_landmarks_segment(video_path, start_s, end_s, target_fps=15, cache_dir=CACHE_DIR):
    key = os.path.basename(video_path) + f"_{start_s:.2f}_{end_s:.2f}_{target_fps}.npz"
    cpath = os.path.join(cache_dir, key)
    if os.path.exists(cpath):
        arr = np.load(cpath)["data"]
        return pd.DataFrame(arr)

    cap = cv2.VideoCapture(video_path)
    orig_fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    start_f = int(start_s * orig_fps)
    end_f   = int(end_s   * orig_fps)
    step = max(1, int(round(orig_fps / target_fps)))

    cap.set(cv2.CAP_PROP_POS_FRAMES, start_f)
    data, fidx = [], start_f
    while cap.isOpened() and fidx < end_f:
        ret, frame = cap.read()
        if not ret: break
        res = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        if res.pose_landmarks:
            row = []
            for lm in res.pose_landmarks.landmark:
                row.extend([lm.x, lm.y, lm.z])
            data.append(row)
        fidx += step
        cap.set(cv2.CAP_PROP_POS_FRAMES, fidx)
    cap.release()

    arr = np.array(data, dtype=np.float32)
    np.savez_compressed(cpath, data=arr)
    return pd.DataFrame(arr)
````

### **C6 — Guardado de features**

````python
# Normaliza por distancia hombro-izq (11) / hombro-der (12)
def normalize_landmarks(df):
    """
    Recibe un DataFrame con los 33 landmarks (x,y,z) de MediaPipe
    y divide cada frame por la distancia entre los hombros para
    quitar escala/distancia a la cámara.
    """
    if df.empty:
        return df
    li = df.iloc[:, 11*3:11*3+3].values   # hombro izq (11)
    ld = df.iloc[:, 12*3:12*3+3].values   # hombro der (12)
    dist = np.linalg.norm(li - ld, axis=1)
    dist[dist == 0] = 1.0   # evitar división por cero

    return df.div(dist, axis=0)


def compute_features(df_norm):
    """
    Saca varios features pensados para tus 5 clases:
    - caminar hacia al frente
    - caminar hacia atras
    - giro 180
    - ponerse de pie
    - sentarse
    """
    if df_norm.empty or len(df_norm) < 2:
        return {}

    n_frames = len(df_norm)
    diffs = df_norm.diff().fillna(0).values
    speeds = np.linalg.norm(diffs, axis=1)

    mean_speed = float(speeds.mean())
    std_speed = float(speeds.std())
    p75_speed = float(np.percentile(speeds, 75))
    p90_speed = float(np.percentile(speeds, 90))

    def mid(a, b):
        return (a + b) / 2.0
    try:
        # secuencias para poder medir variación
        ls_seq = df_norm.iloc[:, 11*3:11*3+2].values   # hombro izq
        rs_seq = df_norm.iloc[:, 12*3:12*3+2].values   # hombro der
        lh_seq = df_norm.iloc[:, 23*3:23*3+2].values   # cadera izq
        rh_seq = df_norm.iloc[:, 24*3:24*3+2].values   # cadera der

        trunk_angles = []
        for i in range(n_frames):
            shoulder_mid = mid(ls_seq[i], rs_seq[i])
            hip_mid = mid(lh_seq[i], rh_seq[i])
            v = shoulder_mid - hip_mid
            trunk_angles.append(np.degrees(np.arctan2(v[1], v[0])))

        trunk_angle_deg = float(np.mean(trunk_angles))   # ángulo medio
        trunk_angle_var = float(np.var(trunk_angles))    # cuánto cambió → giro
    except Exception:
        trunk_angle_deg = 0.0
        trunk_angle_var = 0.0

    # ----- altura cabeza/cadera (para sentarse / ponerse de pie) -----
    # cabeza = 0, cadera izq = 23
    head_y = df_norm.iloc[:, 0*3 + 1].values
    hip_y = df_norm.iloc[:, 23*3 + 1].values

    head_y_mean = float(head_y.mean())
    head_y_min = float(head_y.min())
    head_y_range = float(head_y.max() - head_y.min())
    hip_y_mean = float(hip_y.mean())

    # ----- movimiento en X (para frente vs atrás) -----
    hip_x = df_norm.iloc[:, 23*3 + 0].values
    hip_x_disp = float(hip_x[-1] - hip_x[0])              # desplazamiento neto
    hip_x_path = float(np.abs(np.diff(hip_x)).sum())      # camino total-
    seg_len = float(n_frames)

    return {
        "mean_speed": mean_speed,
        "std_speed": std_speed,
        "p75_speed": p75_speed,
        "p90_speed": p90_speed,
        "trunk_angle_deg": trunk_angle_deg,
        "trunk_angle_var": trunk_angle_var,
        "head_y_mean": head_y_mean,
        "head_y_min": head_y_min,
        "head_y_range": head_y_range,
        "hip_y_mean": hip_y_mean,
        "hip_x_disp": hip_x_disp,
        "hip_x_path": hip_x_path,
        "seg_len": seg_len,
    }

rows = []

for _, r in tqdm(df_segments.iterrows(), total=len(df_segments)):
    df_lm = extract_landmarks_segment(
        r["video_path"],
        r["start_s"],
        r["end_s"]
    )

    # normalizo por hombros
    df_norm = normalize_landmarks(df_lm)
    feats = compute_features(df_norm)
    feats.update({
        "label": r["label"],
        "file": r["file"],
        "start_s": r["start_s"],
        "end_s": r["end_s"],
        "dur_seg": r["dur_seg"],
    })

    rows.append(feats)

df_features = pd.DataFrame(rows)
out_features = f"{ARTIFACTS_DIR}/features.csv"
df_features.to_csv(out_features, index=False)
print("Features guardadas en:", out_features)
df_features.head()
````

## Validación Cruzada para Evaluación Robusta

Implementar un esquema de validación cruzada (Stratified K-Fold Cross-Validation) en el modelo con características reducidas para obtener una estimación más robusta y menos sesgada de su rendimiento general. Se calcularán y promediarán las métricas de clasificación (accuracy, precision, recall, F1-score) a través de los diferentes folds.


````python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd

# Prepare the data with reduced features
X = df_features_eda[selected_features].fillna(0.0)
y = y_enc # y_enc already contains the encoded labels

# 2. Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Initialize RandomForestClassifier with best hyperparameters
clf_cv = RandomForestClassifier(
    n_estimators=100, max_depth=10, min_samples_split=4, random_state=42
)

# 4. Use cross_val_score for accuracy (quick check)
accuracy_scores = cross_val_score(clf_cv, X, y, cv=skf, scoring='accuracy', n_jobs=-1)
print(f"\nCross-validation Accuracy: {accuracy_scores.mean():.2f} \u00b1 {accuracy_scores.std():.2f}")

# 5. Implement manual loop for StratifiedKFold to calculate all metrics
all_accuracy = []
all_precision = []
all_recall = []
all_f1 = []

for train_index, val_index in skf.split(X, y):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y[train_index], y[val_index]

    # b. Train the model on the training set
    clf_cv.fit(X_train, y_train)

    # c. Make predictions on the validation set
    y_pred = clf_cv.predict(X_val)

    # d. Calculate and store metrics for each fold
    all_accuracy.append(accuracy_score(y_val, y_pred))
    all_precision.append(precision_score(y_val, y_pred, average='macro', zero_division=0))
    all_recall.append(recall_score(y_val, y_pred, average='macro', zero_division=0))
    all_f1.append(f1_score(y_val, y_pred, average='macro', zero_division=0))

# 6. Calculate the mean and standard deviation of all metrics
mean_accuracy = np.mean(all_accuracy)
std_accuracy = np.std(all_accuracy)

mean_precision = np.mean(all_precision)
std_precision = np.std(all_precision)

mean_recall = np.mean(all_recall)
std_recall = np.std(all_recall)

mean_f1 = np.mean(all_f1)
std_f1 = np.std(all_f1)

# 7. Print the mean and standard deviation of each metric
print("\n--- Evaluación Robusta con Validación Cruzada (Stratified K-Fold) ---")
print(f"Accuracy:  {mean_accuracy:.4f} \u00b1 {std_accuracy:.4f}")
print(f"Precision: {mean_precision:.4f} \u00b1 {std_precision:.4f}")
print(f"Recall:    {mean_recall:.4f} \u00b1 {std_recall:.4f}")
print(f"F1-Score:  {mean_f1:.4f} \u00b1 {std_f1:.4f}")
````

````python
def get_landmark_coords(landmarks_df, frame_idx, landmark_idx, axis):
    """Obtiene la coordenada (x, y, o z) de un landmark específico en un frame dado."""
    col_idx = landmark_idx * 3 + axis  # 0=x, 1=y, 2=z
    if frame_idx < len(landmarks_df) and col_idx < len(landmarks_df.columns):
        return landmarks_df.iloc[frame_idx, col_idx]
    return None

def calculate_angle(p1, p2, p3):
    """Calcula el ángulo entre tres puntos (p1-p2-p3) en 2D."""
    v1 = np.array([p1[0] - p2[0], p1[1] - p2[1]])
    v2 = np.array([p3[0] - p2[0], p3[1] - p2[1]])

    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)

    if norm_v1 == 0 or norm_v2 == 0:
        return 0.0 # Evitar divisiones por cero

    angle_rad = np.arccos(np.clip(dot_product / (norm_v1 * norm_v2), -1.0, 1.0))
    return np.degrees(angle_rad)

def compute_advanced_features(df_norm, orig_fps):
    """
    Calcula un conjunto más avanzado de características biomecánicas
    incluyendo velocidad, aceleración y ángulos articulares dinámicos.
    df_norm: DataFrame de landmarks normalizados.
    orig_fps: FPS original del video para cálculos de tiempo.
    """
    if df_norm.empty or len(df_norm) < 3: # Necesitamos al menos 3 frames para aceleración
        return {
            "mean_accel": 0.0,
            "std_accel": 0.0,
            "shoulder_hip_angle_mean": 0.0,
            "shoulder_hip_angle_std": 0.0,
            "hip_knee_angle_mean": 0.0,
            "hip_knee_angle_std": 0.0,
            "left_right_hip_asymmetry": 0.0,
            "left_right_shoulder_asymmetry": 0.0,
            "total_motion_mag": 0.0
        }

    n_frames = len(df_norm)
    frame_times = np.arange(n_frames) / orig_fps

    # === Velocidad y Aceleración ===
    # Velocidad de los landmarks (magnitud del cambio entre frames)
    velocities = np.linalg.norm(df_norm.diff().fillna(0).values, axis=1)
    # Aceleración (cambio de velocidad)
    accelerations = np.diff(velocities) / (1 / orig_fps) # Aceleración = diff(velocidad) / diff(tiempo)
    if len(accelerations) == 0:
      mean_accel = 0.0
      std_accel = 0.0
    else:
      mean_accel = np.mean(np.abs(accelerations)) # Magnitud de aceleración promedio
      std_accel = np.std(np.abs(accelerations))

    # === Ángulos Articulares Dinámicos ===
    shoulder_hip_angles = []
    hip_knee_angles = []
    for i in range(n_frames):
        # Puntos para el ángulo hombro-cadera (vista lateral, si es posible)
        # Consideremos un eje en el plano Y (vertical) y Z (profundidad) para un movimiento de lado
        # Para este ejemplo, usaremos coordenadas 2D (X,Y) proyectadas

        # Hombro izquierdo (11), Cadera izquierda (23), Hombro derecho (12), Cadera derecha (24)
        # Rodilla izquierda (25), Rodilla derecha (26)

        # Ángulo Hombro-Cadera (aproximación con puntos medios del tronco)
        ls = df_norm.iloc[i, 11*3:11*3+2].values
        rs = df_norm.iloc[i, 12*3:12*3+2].values
        lh = df_norm.iloc[i, 23*3:23*3+2].values
        rh = df_norm.iloc[i, 24*3:24*3+2].values
        mid_shoulder = (ls + rs) / 2
        mid_hip = (lh + rh) / 2
        # Usamos un punto "virtual" arriba del hombro para el cálculo del ángulo vertical
        virtual_top = mid_shoulder + np.array([0, -1]) # Un punto directamente arriba
        shoulder_hip_angles.append(calculate_angle(virtual_top, mid_shoulder, mid_hip))

        # Ángulo Cadera-Rodilla (ej. izquierda)
        lk = df_norm.iloc[i, 25*3:25*3+2].values
        if not np.isnan(lk).any(): # Solo si la rodilla es visible
             hip_knee_angles.append(calculate_angle(mid_hip, lh, lk))

    shoulder_hip_angle_mean = np.mean(shoulder_hip_angles) if shoulder_hip_angles else 0.0
    shoulder_hip_angle_std = np.std(shoulder_hip_angles) if shoulder_hip_angles else 0.0

    hip_knee_angle_mean = np.mean(hip_knee_angles) if hip_knee_angles else 0.0
    hip_knee_angle_std = np.std(hip_knee_angles) if hip_knee_angles else 0.0

    # === Simetría ===
    # Asimetría de cadera (diferencia en Y)
    left_hip_y = df_norm.iloc[:, 23*3+1].values
    right_hip_y = df_norm.iloc[:, 24*3+1].values
    left_right_hip_asymmetry = np.mean(np.abs(left_hip_y - right_hip_y)) if len(left_hip_y) > 0 else 0.0

    # Asimetría de hombro (diferencia en Y)
    left_shoulder_y = df_norm.iloc[:, 11*3+1].values
    right_shoulder_y = df_norm.iloc[:, 12*3+1].values
    left_right_shoulder_asymmetry = np.mean(np.abs(left_shoulder_y - right_shoulder_y)) if len(left_shoulder_y) > 0 else 0.0

    # === Magnitud total de movimiento (varianza de las coordenadas normalizadas) ===
    total_motion_mag = np.var(df_norm.values) # Varianza de todas las coordenadas


    return {
        "mean_accel": float(mean_accel),
        "std_accel": float(std_accel),
        "shoulder_hip_angle_mean": float(shoulder_hip_angle_mean),
        "shoulder_hip_angle_std": float(shoulder_hip_angle_std),
        "hip_knee_angle_mean": float(hip_knee_angle_mean),
        "hip_knee_angle_std": float(hip_knee_angle_std),
        "left_right_hip_asymmetry": float(left_right_hip_asymmetry),
        "left_right_shoulder_asymmetry": float(left_right_shoulder_asymmetry),
        "total_motion_mag": float(total_motion_mag)
    }
````

## Manejo del Desbalance de Clases (SMOTE + XGBoost)

Implementar la técnica de sobremuestreo SMOTE en el conjunto de entrenamiento de XGBoost para balancear las clases y entrenar el modelo, con el objetivo de mejorar el rendimiento en clases minoritarias.


````python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

print("Starting XGBoost with SMOTE model training and evaluation...")

# --- ALIAS CANÓNICO DEL OBJETIVO (refined -> advanced) ---
y_enc_advanced = y_enc_refined  # usar el target que verdaderamente existe en tu flujo
le_advanced = le_refined

# 1. Divida el DataFrame X_advanced y el array y_enc_advanced en conjuntos de entrenamiento y prueba
# X_advanced and y_enc_advanced are already available from previous cells.
X_train, X_test, y_train, y_test = train_test_split(
    X_advanced, y_enc_advanced, test_size=0.25, random_state=42, stratify=y_enc_advanced
)

# 2. Inicialice un objeto SMOTE y aplíquelo únicamente al conjunto de entrenamiento
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train, y_train)

print(f"\nOriginal training set shape: {X_train.shape}, {y_train.shape}")
print(f"Resampled training set shape: {X_resampled.shape}, {y_resampled.shape}")
print("Class distribution after SMOTE:", pd.Series(y_resampled).value_counts())

# 3. Inicialice un xgb.XGBClassifier con los mejores hiperparámetros y entrene con los datos balanceados
# gs_xgb.best_params_ was found in the previous step (de455b32)
# Make sure to handle 'use_label_encoder' if it's in best_params
best_params_xgb = gs_xgb.best_params_.copy()
if 'use_label_encoder' in best_params_xgb:
    del best_params_xgb['use_label_encoder']  # Suppress warning

clf_xgb_smote = xgb.XGBClassifier(**best_params_xgb, random_state=42)
clf_xgb_smote.fit(X_resampled, y_resampled)

# 4. Realice predicciones sobre el conjunto de prueba
y_pred_xgb_smote = clf_xgb_smote.predict(X_test)

# 5. Imprima el classification_report y la confusion_matrix
print("\nClassification Report para XGBoost con SMOTE (conjunto de prueba):")
print(classification_report(y_test, y_pred_xgb_smote, target_names=le_advanced.classes_))

print("\nMatriz de Confusión para XGBoost con SMOTE (conjunto de prueba):\n")
print(confusion_matrix(y_test, y_pred_xgb_smote))
````

## Evaluación del Modelo XGBoost con SMOTE

Evaluar el rendimiento del modelo XGBoost entrenado con SMOTE mediante validación cruzada. Esto incluirá la generación de un `classification_report` y una `confusion_matrix` en el conjunto de prueba, así como una validación cruzada robusta para obtener métricas promedio (accuracy, precision, recall, F1-score).


````python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

print("Starting robust evaluation with Stratified K-Fold Cross-Validation for XGBoost with SMOTE...")

# 1. Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 2. Initialize xgb.XGBClassifier with the best hyperparameters
# best_params_xgb is already defined from previous steps
clf_xgb_smote_cv = xgb.XGBClassifier(**best_params_xgb, random_state=42)

# Store metrics for each fold
all_accuracy_xgb_smote = []
all_precision_xgb_smote = []
all_recall_xgb_smote = []
all_f1_xgb_smote = []

# 3. Perform manual loop for StratifiedKFold cross-validation
for train_index, val_index in skf.split(X_advanced, y_enc_advanced):
    X_train_fold, X_val_fold = X_advanced.iloc[train_index], X_advanced.iloc[val_index]
    y_train_fold, y_val_fold = y_enc_advanced[train_index], y_enc_advanced[val_index]

    # b. Apply SMOTE to the training set only
    sm_cv = SMOTE(random_state=42)
    X_resampled_fold, y_resampled_fold = sm_cv.fit_resample(X_train_fold, y_train_fold)

    # c. Train the xgb.XGBClassifier on the resampled training data
    clf_xgb_smote_cv.fit(X_resampled_fold, y_resampled_fold)

    # d. Make predictions on the validation set
    y_pred_fold = clf_xgb_smote_cv.predict(X_val_fold)

    # e. Calculate and store metrics for each fold
    all_accuracy_xgb_smote.append(accuracy_score(y_val_fold, y_pred_fold))
    all_precision_xgb_smote.append(precision_score(y_val_fold, y_pred_fold, average='macro', zero_division=0))
    all_recall_xgb_smote.append(recall_score(y_val_fold, y_pred_fold, average='macro', zero_division=0))
    all_f1_xgb_smote.append(f1_score(y_val_fold, y_pred_fold, average='macro', zero_division=0))

# 4. Calculate and display the mean and standard deviation of each metric
mean_accuracy_xgb_smote = np.mean(all_accuracy_xgb_smote)
std_accuracy_xgb_smote = np.std(all_accuracy_xgb_smote)

mean_precision_xgb_smote = np.mean(all_precision_xgb_smote)
std_precision_xgb_smote = np.std(all_precision_xgb_smote)

mean_recall_xgb_smote = np.mean(all_recall_xgb_smote)
std_recall_xgb_smote = np.std(all_recall_xgb_smote)

mean_f1_xgb_smote = np.mean(all_f1_xgb_smote)
std_f1_xgb_smote = np.std(all_f1_xgb_smote)

print("\n--- Evaluación Robusta con Validación Cruzada (Stratified K-Fold) para XGBoost con SMOTE ---")
print(f"Accuracy:  {mean_accuracy_xgb_smote:.4f} \u00b1 {std_accuracy_xgb_smote:.4f}")
print(f"Precision: {mean_precision_xgb_smote:.4f} \u00b1 {std_precision_xgb_smote:.4f}")
print(f"Recall:    {mean_recall_xgb_smote:.4f} \u00b1 {std_recall_xgb_smote:.4f}")
print(f"F1-Score:  {mean_f1_xgb_smote:.4f} \u00b1 {std_f1_xgb_smote:.4f}")
````

## Revisión y Refinamiento de Características Clave para Clases Problemáticas

Refinar y agregar características biomecánicas enfocadas en mejorar la discriminación de las actividades 'caminar hacia al frente' y 'sentarse', y luego re-extraer todas las características para el conjunto de datos.


````python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False,
                    model_complexity=1,
                    min_detection_confidence=0.5,
                    min_tracking_confidence=0.5)

def extract_landmarks_segment(video_path, start_s, end_s, target_fps=15, cache_dir=CACHE_DIR):
    key = os.path.basename(video_path) + f"_{start_s:.2f}_{end_s:.2f}_{target_fps}.npz"
    cpath = os.path.join(cache_dir, key)
    if os.path.exists(cpath):
        arr = np.load(cpath)["data"]
        return pd.DataFrame(arr)

    cap = cv2.VideoCapture(video_path)
    orig_fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
    start_f = int(start_s * orig_fps)
    end_f   = int(end_s   * orig_fps)
    step = max(1, int(round(orig_fps / target_fps)))

    cap.set(cv2.CAP_PROP_POS_FRAMES, start_f)
    data, fidx = [], start_f
    while cap.isOpened() and fidx < end_f:
        ret, frame = cap.read()
        if not ret: break
        res = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        if res.pose_landmarks:
            row = []
            for lm in res.pose_landmarks.landmark:
                row.extend([lm.x, lm.y, lm.z])
            data.append(row)
        fidx += step
        cap.set(cv2.CAP_PROP_POS_FRAMES, fidx)
    cap.release()

    arr = np.array(data, dtype=np.float32)
    np.savez_compressed(cpath, data=arr)
    return pd.DataFrame(arr)

# Normaliza por distancia hombro-izq (11) / hombro-der (12)
def normalize_landmarks(df):
    """
    Recibe un DataFrame con los 33 landmarks (x,y,z) de MediaPipe
    y divide cada frame por la distancia entre los hombros para
    quitar escala/distancia a la cámara.
    """
    if df.empty:
        return df
    li = df.iloc[:, 11*3:11*3+3].values   # hombro izq (11)
    ld = df.iloc[:, 12*3:12*3+3].values   # hombro der (12)
    dist = np.linalg.norm(li - ld, axis=1)
    dist[dist == 0] = 1.0   # evitar división por cero

    return df.div(dist.reshape(-1,1), axis=0)

def compute_features(df_norm, orig_fps):
    """
    Saca varios features pensados para tus 5 clases:
    - caminar hacia al frente
    - caminar hacia atras
    - giro 180
    - ponerse de pie
    - sentarse

    Incluye nuevas características para mejorar la discriminación.
    """
    if df_norm.empty or len(df_norm) < 2:
        return {}

    n_frames = len(df_norm)
    diffs = df_norm.diff().fillna(0).values
    speeds = np.linalg.norm(diffs, axis=1)

    mean_speed = float(speeds.mean())
    std_speed = float(speeds.std())
    p75_speed = float(np.percentile(speeds, 75))
    p90_speed = float(np.percentile(speeds, 90))

    def mid(a, b):
        return (a + b) / 2.0
    try:
        # secuencias para poder medir variación
        ls_seq = df_norm.iloc[:, 11*3:11*3+2].values   # hombro izq (x,y)
        rs_seq = df_norm.iloc[:, 12*3:12*3+2].values   # hombro der (x,y)
        lh_seq = df_norm.iloc[:, 23*3:23*3+2].values   # cadera izq (x,y)
        rh_seq = df_norm.iloc[:, 24*3:24*3+2].values   # cadera der (x,y)

        trunk_angles = []
        for i in range(n_frames):
            shoulder_mid = mid(ls_seq[i], rs_seq[i])
            hip_mid = mid(lh_seq[i], rh_seq[i])
            v = shoulder_mid - hip_mid
            # We use a placeholder for calculation that was previously done with 2D vectors
            # For a direct calculation of trunk angle with respect to vertical, we need 3D or better context
            # Keeping as it was, but this is an approximation.
            trunk_angles.append(np.arctan2(v[1], v[0])) # angle in radians

        trunk_angle_deg = float(np.degrees(np.mean(trunk_angles))) # ángulo medio
        trunk_angle_var = float(np.var(trunk_angles)) # cuánto cambió → giro
    except Exception:
        trunk_angle_deg = 0.0
        trunk_angle_var = 0.0

    # ----- altura cabeza/cadera (para sentarse / ponerse de pie) -----
    # cabeza = 0, cadera izq = 23
    head_y = df_norm.iloc[:, 0*3 + 1].values # Y-coordinate of head
    hip_y = df_norm.iloc[:, 23*3 + 1].values # Y-coordinate of left hip

    head_y_mean = float(head_y.mean())
    head_y_min = float(head_y.min())
    head_y_range = float(head_y.max() - head_y.min())
    hip_y_mean = float(hip_y.mean())

    # ----- movimiento en X (para frente vs atrás) -----
    hip_x = df_norm.iloc[:, 23*3 + 0].values # X-coordinate of left hip
    hip_x_disp = float(hip_x[-1] - hip_x[0])              # desplazamiento neto
    hip_x_path = float(np.abs(np.diff(hip_x)).sum())      # camino total
    seg_len = float(n_frames)

    # --- Nuevas características ---
    # 1. avg_vertical_speed_hip: Velocidad vertical promedio absoluta de la cadera
    hip_y_diff = np.diff(hip_y)
    avg_vertical_speed_hip = float(np.mean(np.abs(hip_y_diff))) if len(hip_y_diff) > 0 else 0.0

    # 2. avg_horizontal_speed_hip: Velocidad horizontal promedio absoluta de la cadera
    hip_x_diff = np.diff(hip_x)
    avg_horizontal_speed_hip = float(np.mean(np.abs(hip_x_diff))) if len(hip_x_diff) > 0 else 0.0

    # 3. normalized_head_hip_distance_mean: Distancia vertical promedio entre cabeza y cadera, normalizada
    # Recalculate shoulder distance for normalization within this function if needed, or assume df_norm is already normalized correctly for positions
    # If df_norm is normalized by shoulder width, then distance directly between landmarks is also normalized.
    head_hip_distance = head_y - hip_y
    normalized_head_hip_distance_mean = float(np.mean(head_hip_distance)) if len(head_hip_distance) > 0 else 0.0

    return {
        "mean_speed": mean_speed,
        "std_speed": std_speed,
        "p75_speed": p75_speed,
        "p90_speed": p90_speed,
        "trunk_angle_deg": trunk_angle_deg,
        "trunk_angle_var": trunk_angle_var,
        "head_y_mean": head_y_mean,
        "head_y_min": head_y_min,
        "head_y_range": head_y_range,
        "hip_y_mean": hip_y_mean,
        "hip_x_disp": hip_x_disp,
        "hip_x_path": hip_x_path,
        "seg_len": seg_len,
        "avg_vertical_speed_hip": avg_vertical_speed_hip,
        "avg_horizontal_speed_hip": avg_horizontal_speed_hip,
        "normalized_head_hip_distance_mean": normalized_head_hip_distance_mean
    }


rows_refined = []

for _, r in tqdm(df_segments.iterrows(), total=len(df_segments)):
    df_lm = extract_landmarks_segment(
        r["video_path"],
        r["start_s"],
        r["end_s"]
    )

    # normalizo por hombros
    df_norm = normalize_landmarks(df_lm)
    # Pass orig_fps to compute_features for potential time-dependent calculations
    feats_refined = compute_features(df_norm, r["fps"])
    feats_refined.update({
        "label": r["label"],
        "file": r["file"],
        "start_s": r["start_s"],
        "end_s": r["end_s"],
        "dur_seg": r["dur_seg"],
    })

    rows_refined.append(feats_refined)

df_features_refined = pd.DataFrame(rows_refined)
out_features_refined = f"{ARTIFACTS_DIR}/features_refined.csv"
df_features_refined.to_csv(out_features_refined, index=False)
print("Features refinadas guardadas en:", out_features_refined)
display(df_features_refined.head())
````

El modelo, al clasificar con precisión la mayoría de las actividades, ofrece un marco para:
